```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import WebBaseLoader
from langchain_community.vectorstores import Chroma
from langchain_openai import OpenAIEmbeddings

docs = [WebBaseLoader(url).load() for url in urls]
docs_list = [item for sublist in docs for item in sublist]

text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=250, chunk_overlap=0
)
doc_splits = text_splitter.split_documents(docs_list)


# Add to vectorDB
vectorstore = Chroma.from_documents(
    documents=doc_splits,
    collection_name="rag-chroma",
    embedding=OpenAIEmbeddings(),
)
retriever = vectorstore.as_retriever(search_type="similarity_score_threshold", search_kwargs={"score_threshold": 0.5})
```

```python
retriever.invoke("what is the best place to visit in benin for eco tourism?")
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough, RunnableMap
from langchain_core.prompts import PromptTemplate

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

# Assuming retriever and llm are properly defined elsewhere
context = retriever | format_docs
question = "Make me a complete travel guide for eco-tourism in benin and tell me what places should I visit. My budget in Benin is $2000 "

# Define the chain of operations
rag_chain = (
    RunnableMap({"context": context , "question": RunnablePassthrough()})
    | prompt
    | llm
    | StrOutputParser()
)

# Stream the results
for chunk in rag_chain.stream({"context": context, "question": question}):
    print(chunk, end="", flush=True)
```

```python
loader = AsyncHtmlLoader(urls)
docs = loader.load()
```

```python
from langchain_community.document_transformers import BeautifulSoupTransformer
# Transform
bs_transformer = BeautifulSoupTransformer()
docs_transformed = bs_transformer.transform_documents(docs,tags_to_extract=["span","p","li","a"])
```

```python
# Result
docs_transformed
```

```python
import os
from langchain.chains import create_extraction_chain
from langchain.llms import OpenAI
from langchain_text_splitters import RecursiveCharacterTextSplitter



# Initialize the LLM
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(temperature=0, model="gpt-4o")

# Define the schema for extraction
schema = {
    "properties": {
        "tourist_places_in_Benin_for_eco_tourism": {"type": "string"},
        "place_description": {"type": "string"},
        "planning_tip": {"type": "string"},
        "best_time_to_visit": {"type": "string"},
        "budget": {"type": "string"},
    },
    "required": ["places_for_tourist_for_eco_tourism", "place_description","planning_tip","best_time_to_visit","budget"],
}

# Function to perform extraction
def extract(content: str, schema: dict):
    chain =create_extraction_chain(schema=schema, llm=llm)
    return chain.run(content)
# Initialize the text splitter
splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=1000, chunk_overlap=0
)


# Split the documents into chunks
splits = splitter.split_documents(docs_transformed)

# Extract and print the content for each split
import pprint
import json
extracted_contents =[]
for split in splits:
    extracted_content = extract(schema=schema, content=split.page_content)
    json_content =json.dumps(extracted_content)
    extracted_contents.append(extracted_content)
    pprint.pprint(extracted_content)

with open('extracted_content.json', 'w') as json_file:
    json.dump(extracted_contents, json_file, indent=4)

print(extracted_contents)


```
